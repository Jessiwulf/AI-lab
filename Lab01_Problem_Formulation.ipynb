## Conditionals

```python
my_variable = 5
if my_variable < 0:
  print("negative")
elif my_variable == 0:
  print("null")
else: # my_variable > 0
  print("positive")
```

**Exercise 7.1 [5 points].** Using a conditional, write a function to check whether a given number is even or odd





```python
# Global variable to store even/odd result
last_check = None

def even_odd(x, silent=False):
  # Write your function here
    global last_check
    if x % 2 == 0:
        last_check = "even"
        if not silent:
            print(f"{x} is an even integer")
    else:
        last_check = "odd"
        if not silent:
            print(f"{x} is an odd integer")

even_odd(-3)
even_odd(20)

# Expected output:
# -3 is an odd integer
# 20 is an even integer
```

Reuse the *even_odd* function above.

```python
def even_distance(list):
  # Write your function here
    even_indices = []

    for i in range(len(list)):
        even_odd(list[i], silent=True)  # Suppress print
        if last_check == "even":
            even_indices.append(i)

    if len(even_indices) >= 2:
        distance = even_indices[-1] - even_indices[0]
        return distance
    else:
        return 0

test_list = [1,3,2,9,7,4,0,8,4,10]
even_distance(test_list)

# Expected output:
# 7
# note: Even numbers begin at '2' and end at '10', index distance is 7 indices
```

**Exercise 7.4 [5 points]** write a function that sorts a list in ascending order (from smaller to bigger) using the [bubble sort](https://en.wikipedia.org/wiki/Bubble_sort) algorithm.

```python
def bubble_sort(my_list):
  # Write your function here
  n = len(my_list)
  for i in range(n):
      for j in range(0, n - 1 - i):
          if my_list[j] > my_list[j + 1]:
              # Swap
              my_list[j], my_list[j + 1] = my_list[j + 1], my_list[j]
  return my_list

my_list = [1, -3, 3, 2]
bubble_sort(my_list)

# Should return [-3, 1, 2, 3]
```

**2.7 [5 points]** What is the **final state**?

Your Answer:

The final state is when all waste items presented to the robot have been successfully processed and placed into their correct, designated sorted bins (e.g., "Recyclable Plastics," "Recyclable Metals," "General Waste," "Organic Waste," etc.).

**2.8 [5 points]** What are the **actions** that made states change?

Your Answer:

- Sense/Perceive: The robot uses its sensors to analyze an item (e.g., capturing images, measuring weight, scanning for metal). This action changes the robot's internal "knowledge state" about the item.

- Identify/Classify: Based on sensor data, the AI processes the information to classify the item (e.g., "plastic bottle," "glass jar," "food waste"). This is a cognitive action that doesn't physically change the item's location but changes its classification within the system.

- Grasp/Pick Up: The robotic arm extends, activates its gripper/suction cup, and picks up the identified waste item. This changes the item's location from the input area to being held by the robot.

- Move/Transport: The robotic arm moves the item from the pick-up location to the appropriate disposal bin. This is a physical change of the item's position.

- Release/Deposit: The robotic arm releases the item into the correct bin (e.g., recyclable bin, non-recyclable bin). This action changes the item's state from "being held" to "being in a sorted bin."
